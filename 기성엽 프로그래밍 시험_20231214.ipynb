```python
# 파일 열기
df_lib_2020 = pd.read_excel("성동구립도서관 장서 대출목록 (2020년 11월).xlsx")
df_lib_2023 = pd.read_excel("성동구립도서관 장서 대출목록 (2023년 11월).xlsx")
print(df_lib_2020)
print(df_lib_2023)
```

1. 2020년 11월과 비교하여 2023년 11월 시점에서 추가된 도서의 목록 및 개수

```python
df_lib_2023[~df_lib_2023['ISBN'].isin(df_lib_2020['ISBN'])] # ISBN일 기준으로 23년에 추가된 도서리스트 추출
# 36419 권
```

2. 2020년 11월과 비교하여 2023년 11월 시점에서 없어진 도서의 목록 및 개수

```python
df_lib_2020[~df_lib_2020['ISBN'].isin(df_lib_2023['ISBN'])] # ISBN일 기준으로 20년에 포함되지 않은 도서리스트 추출
# 28840 권
```

3. 2020년 11월과 2023년 11월 가장 대출이 많이 발생한 도서 상위 20권의 리스트와 그 빈도수 비교

```python
# 대출 상위 20권 리스트 추출
df_lib_2020_rend_top20 = df_lib_2020.sort_values('대출건수', ascending=False).head(20).reset_index(drop=True)
df_lib_2023_rend_top20 = df_lib_2023.sort_values('대출건수', ascending=False).head(20).reset_index(drop=True)
# 빈도수(대출건수 합계) 비교
print('2020년 11월 대출 상위 20권 도서 목록과 빈도수(대출건수 합계)')
display(df_lib_2020_rend_top20)
print('대출건수 합계: ', df_lib_2020_rend_top20['대출건수'].sum(), '권')
print('2023년 11월 대출 상위 20권 도서 목록과 빈도수(대출건수 합계)')
display(df_lib_2023_rend_top20)
print('대출건수 합계: ', df_lib_2023_rend_top20['대출건수'].sum(), '권')
```

4. 2020년 11월과 2023년 11월 가장 대출이 많이 발생한 도서 상위 100권 중 동일한 도서의 리스트와 그 개수 (100권이 적당하지 않으면 숫자를 조정 가능)

```python
# 대출 상위 100권 리스트 추출
df_lib_2020_rend_top100 = df_lib_2020.sort_values('대출건수', ascending=False).head(100).reset_index(drop=True)
df_lib_2023_rend_top100 = df_lib_2023.sort_values('대출건수', ascending=False).head(100).reset_index(drop=True)
# 동일한 도서와 개수
df_equal_books = df_lib_2023_rend_top100[df_lib_2023_rend_top100['ISBN'].isin(df_lib_2020_rend_top100['ISBN'])]
display(df_lib_2023_rend_top100[df_lib_2023_rend_top100['ISBN'].isin(df_lib_2020_rend_top100['ISBN'])])
print('동일한 도서 개수: ', df_equal_books.shape[0], '권')
```

5. 2020년 11월 가장 대출이 많이 발생한 도서 상위 50권의 당시 대출 횟수와 현재 대출 횟수 비교하여 증가한 도서와 감소한 도서 파악, 그래프로 표현

```python
df_lib_2020_rend_top50 = df_lib_2020.sort_values('대출건수', ascending=False).head(50).reset_index(drop=True)
# 현재 대출 회수 결합
df_lib_2023_rent_v2 = df_lib_2023[['도서명', 'ISBN', '발행년도', '대출건수']].groupby(['도서명', 'ISBN', '발행년도'], as_index=False)['대출건수'].sum() # 중복된 목록에서 대출건수 합계 생성
df_lib_2020_rend_top50_2023_rent_cnt = df_lib_2020_rend_top50.merge(df_lib_2023_rent_v2, how='left', on=['도서명', '발행년도', 'ISBN'])
df_lib_2020_rend_top50_2023_rent_cnt.rename({'대출건수_x': '대출건수_2020년', '대출건수_y': '대출건수_2023년'}, axis=1, inplace=True)
df_lib_2020_rend_top50_2023_rent_cnt
# 대출 증가 도서
df_lib_2020_rend_top50_2023_rent_cnt['대출차이'] = df_lib_2020_rend_top50_2023_rent_cnt['대출건수_2023년'] - df_lib_2020_rend_top50_2023_rent_cnt['대출건수_2020년']

# 시각화
df_plot = df_lib_2020_rend_top50_2023_rent_cnt[['도서명', '대출건수_2020년', '대출건수_2023년']].copy()
df_plot['대출건수_2023년'].fillna(0, inplace=True)
## 그래프 그리기
plt.figure(figsize=(20,6))
bar_width = 0.35
x = range(df_plot['도서명'].shape[0])
## bar 그래프 그리기
plt.bar(x, df_plot['대출건수_2020년'], width=bar_width, label='대출건수_2020년', color='b')
plt.bar([p + bar_width for p in x], df_plot['대출건수_2023년'], width=bar_width, label='대출건수_2023년', color='g')
## label 등 설정
plt.xlabel('도서명')
plt.ylabel('대출건수')
plt.title('년도별 대출건수 비교')
plt.xticks([p + bar_width/2 for p in x], df_plot['도서명'], rotation=45, ha='right')
plt.legend()
plt.show()
```
